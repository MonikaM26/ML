## The Data

The data frame consists of  777 observations on the following 18 variables.

* Enroll Number of new students enrolled

* Top25perc Pct. new students from top 25% of H.S. class
* F.Undergrad Number of fulltime undergraduates
* P.Undergrad Number of parttime undergraduates
* Outstate Out-of-state tuition
* Room.Board Room and board costs
* Books Estimated book costs
* Personal Estimated personal spending
* PhD Pct. of faculty with Ph.D.’s
* Terminal Pct. of faculty with terminal degree
* S.F.Ratio Student/faculty ratio
* perc.alumni Pct. alumni who donate
* Expend Instructional expenditure per student
* Grad.Rate Graduation rate

## K Means Cluster Creation

Creating the Cluster labels.

** Importing KMeans from SciKit Learn.**

```python
from sklearn.cluster import KMeans
```

** Creating an instance of a K Means model with 2 clusters.**

```python
kmeans = KMeans(n_clusters=2)
```

Fitting the model to all the data except for the Private label.

```python
file.head()
kmeans.fit(file.drop(['Private'], axis=1))
```

Checking value of the cluster center vectors

```python
kmeans.cluster_centers_
```

## Evaluation

** Creating a new column for df called 'Cluster', which is a 1 for a Private school, and a 0 for a public school.**

```python
d ={'Yes':1, 'No':0}
file['Cluster']= pd.get_dummies(file['Private'],drop_first=True)
print(file['Cluster'])
```

```python
sns.scatterplot(data=file,x='Room.Board', y='Grad.Rate',hue='Private')
plt.show()
sns.scatterplot(data=file,x='Room.Board', y='Grad.Rate',hue=kmeans.labels_)
```

** Creating a confusion matrix and classification report to see how well the Kmeans clustering worked without being given any labels.**

```python
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(file['Cluster'],kmeans.labels_))
print(classification_report(file['Cluster'],kmeans.labels_))
```

The results are not so bad considering the algorithm is purely using the features to cluster the universities into 2 distinct groups.
